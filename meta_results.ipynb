```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from matplotlib.ticker import MultipleLocator

# Load data from Excel file
file_path = "file path/baseline_returns.xlsx"
df1 = pd.read_excel(file_path)

file_path = "file path/meta_returns.xlsx"
df2 = pd.read_excel(file_path)

file_path = "file path/meta_thompson_returns.xlsx"
df3 = pd.read_excel(file_path)
```

```python
y11 = np.round(df1['success1'].values,6)
y12 = np.round(df1['success2'].values,6)
y13 = np.round(df1['success3'].values,6)
y14 = np.round(df1['success4'].values,6)
y15 = np.round(df1['success5'].values,6)

y21 = np.round(df2['success1'].values,6)
y22 = np.round(df2['success2'].values,6)
y23 = np.round(df2['success3'].values,6)
y24 = np.round(df2['success4'].values,6)
y25 = np.round(df2['success5'].values,6)

y31 = np.round(df3['success1'].values,6)
y35 = np.round(df3['success2'].values,6)
y32 = np.round(df3['success3'].values,6)
y33 = np.round(df3['success4'].values,6)
y34 = np.round(df3['success5'].values,6)

mean = []
q1 = []
q3 = []
for j in range(5):
    mean.append([])
    q1.append([])
    q3.append([])
    
    for i in range(len(y31)):
        
        data1 = [y11[i]+1e-5,y12[i]+1e-5,y13[i]+1e-5,y14[i]+1e-5,y15[i]+1e-5]     # for baseline
        data2 = [y21[i]+1e-5,y22[i]+1e-5,y23[i]+1e-5,y24[i]+1e-5,y25[i]+1e-5]     # for MAML
        data3 = [y31[i]+1e-5,y32[i]+1e-5,y33[i]+1e-5,y34[i]+1e-5,y35[i]+1e-5]     # for MAML thompson
        
        data = [data1,data2,data3]
        
        m = np.mean(data[j])
        SE = np.std(data[j])/np.sqrt(5)
        confidence_interval = stats.t.interval(0.95, df=4, loc=m, scale=SE)
        upper_limit = min(confidence_interval[1], 1)
        mean[j].append(m)
        q1[j].append(confidence_interval[0])
        q3[j].append(upper_limit)

x_range = np.arange(len(y11))
plt.figure(figsize=(10, 8))

plt.plot(x_range, mean[0],color = 'red', label='Mean_baseline')
plt.fill_between(x_range, q1[0], q3[0], alpha=0.35,color = 'pink', label='95 % confidence_TD3')

plt.plot(x_range, mean[1],color = 'blue', label='Mean_MAML')
plt.fill_between(x_range, q1[1], q3[1], alpha=0.35,color = 'grey', label='95 % confidence_SAC')

plt.plot(x_range, mean[2],color = 'green', label='Mean_MAML_thompson')
plt.fill_between(x_range, q1[2], q3[2], alpha=0.35,color = 'lightgreen', label='95 % confidence_DDPG')


plt.grid(True)
plt.xlim(0, len(x_range))
plt.ylim(-0.01,1.02)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Number of episodes', fontsize = 21,labelpad = 10)
plt.ylabel('Episodic Returns', fontsize = 21)
plt.legend(loc='lower right', fontsize = 17)
plt.title('episodic returns', fontsize = 22,pad = 10)


tick_positions = np.arange(0, len(original_episodes) + 1, 100) * batch_size
ax2.set_xticks(tick_positions)

tick_labels = [str(i) for i in tick_positions]
ax2.set_xticklabels(tick_labels, fontsize = 16)

plt.tight_layout()
plt.savefig('path/returns.png', dpi=400, bbox_inches='tight')
plt.show()
```
